### Q1. [1 point] The equation of a line is defined by the formula y = mx + b.  Using numpy and matplotlib libraries, create a Python program that utilizes this equation to plot a line given parameters m = 0.5, intercept b = 2.0 and an array of the following 5 x-coordinates as input: [ 1, 2, 3, 4, 5 ].  

````python
#Q1
import numpy as np
import matplotlib.pyplot as plt

inputs = [1,2,3,4,5]

x = np.linspace(1, 5, 100)
y = 0.5 * x +2.0

for px in inputs:
    py = 0.5*px+2.0
    plt.scatter(px,py)

plt.plot(x,y)

plt.show()
````
