```python
#1
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = 0.5 * x +2.0

plt.plot(x,y)
```

```python
#2
from sklearn.datasets.samples_generator import make_blobs

X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=1)
print(X)
print(y)
plt.scatter(X[:,0],X[:,1],c=y)

```

```python
#3
label_color_map = ['red', 'blue']

label_colors = []        

for i in y:
    label_colors.append(label_color_map[i])
    
plt.scatter(X[:,0],X[:,1],c=label_colors)
```

```python
#4
from sklearn import linear_model
Xnew = [[-0.75, 2.0]]

classifier = linear_model.LogisticRegression(solver = 'liblinear')
classifier.fit(X,y)
pred = classifier.predict(Xnew)
print(pred)

```

```python
#4x
X,y= make_blobs(n_samples=10, centers=2, n_features=2, random_state=1)
pred = classifier.predict(X)
print(y)
print(pred)
```

```python
#5
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=2, noise=0.1, random_state=1)
plt.scatter(X[:,0],X[:,1],c=y)
```
