```python
from textgenrnn import textgenrnn
textgen = textgenrnn()
textgen.train_from_file('anime_name.txt',num_epochs = 10)
textgen.save('anime_name_10e.hdf5')
```

```python
textgen_pokemon = textgenrnn('anime_name_10e.hdf5')

#0.2
tg1 = textgen_pokemon.generate(10,temperature = 0.2,return_as_list = True)
#0.5
tg2 = textgen_pokemon.generate(10,temperature = 0.5,return_as_list = True)
#1
tg3 = textgen_pokemon.generate(10,temperature = 1,return_as_list = True)

print(tg1)
print(tg2)
print(tg3)
```
