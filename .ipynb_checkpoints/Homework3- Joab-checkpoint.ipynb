```python
from textgenrnn import textgenrnn
textgen = textgenrnn()
textgen.train_from_file('Fifa19Names.txt',num_epochs = 10)
textgen.save('Fifa19Names_10e.hdf5')
```

```python
textgen_fifa = textgenrnn('Fifa19Names_10e.hdf5')
```
