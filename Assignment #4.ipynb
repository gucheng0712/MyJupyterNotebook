```python
import os

cd = os.getcwd()
print("cd: ", cd)
filepath = os.path.join(cd, 'alice_in_wonderland.txt')
print("filepath: ", filepath)

#  Read text file
file = open(filepath, "r", encoding="utf8")
alice_text = file.read()

# Replace escape character with space
alice_text = alice_text.replace("\n", " ")

# Make all the text lowercase
alice_text = alice_text.lower()

print(alice_text[:885])
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
tokenized_sentences  = sent_tokenize(alice_text)
print("Sentence Tokenization: ", tokenized_sentences[1])
tokenized_words = word_tokenize(tokenized_sentences[1])
print("Word Tokenization: ", tokenized_words)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in tokenized_words]
print(lemmatized_words)
```

```python
from nltk import pos_tag
tagged_words = pos_tag(tokenized_words)
print(tagged_words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
vectorizer  = CountVectorizer()
document_term_matrix= vectorizer.fit_transform(tokenized_words)

bag_of_words = np.array(vectorizer.get_feature_names())
print(bag_of_words)
print("\nDocument term matrix:")
print(document_term_matrix)
for word, item in zip(bag_of_words, document_term_matrix):
    print(word,", ", document_term_matrix)
```
